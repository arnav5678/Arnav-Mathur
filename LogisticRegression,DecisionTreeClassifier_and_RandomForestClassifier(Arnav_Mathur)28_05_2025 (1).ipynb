```python
# LogisticRegression ----> When we have categorical target column

# Backend -----> Activitation function ---> sigmoid ------> Probablity convrt all sub categories ---->
# threshold_value = 0.5

# 0.5<= output = 1
# 0.5> output = 0
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("covid_toy.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
```

```python
df['fever'] = si.fit_transform(df[['fever']])
```

```python
df.isnull().sum()
```

```python
df.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['gender'] = lb.fit_transform(df['gender'])
df['cough'] = lb.fit_transform(df['cough'])
df['city'] = lb.fit_transform(df['city'])
df['has_covid'] = lb.fit_transform(df['has_covid'])
```

```python
df.head(3)
```

```python
x = df.drop(columns =['has_covid'])

y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
# DecisionTreeClassifier ----->
# Changes in entropy and gini index

# -p1log1-p2log2 # if we have 2sub-categories
# -p1log1-p2log2-p3log3..... -pnlogn ### if we have more than 2 sub-categories
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

# **RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

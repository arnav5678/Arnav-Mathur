```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("mall.csv")
```

```python
df.head(3)
```

```python
df = df.drop(columns = ['CustomerID','Genre'])
```

```python
df.head(3)
```

```python
x= df.iloc[:,[0,1]].values
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt

a=[]

for i in range(1,11):
  b=KMeans(n_clusters=i, init ='k-means++', random_state=42)
  b.fit(x)
  a.append(b.inertia_)

plt.plot(range(1,11), a)

plt.title("The Elbow Method Graph")
plt.xlabel("Number of cluster9k")
plt.ylabel('wcss_list')
plt.show()
```

# **From the above plot, we can see the elbow point is at 4. So the number of cluster here will be 4**

```python
b = KMeans(n_clusters=4, init='k-means++',random_state=42)
y_predict = b.fit_predict(x)
```

```python
# Visulazing the clusters
plt.scatter(x[y_predict==0,0],x[y_predict==0,1], s=100,c='blue',label ='Cluster1') # for first cluster
plt.scatter(x[y_predict==1,0],x[y_predict==1,1], s=100,c='green',label ='Cluster2') # for second cluster
plt.scatter(x[y_predict==2,0],x[y_predict==2,1], s=100,c='red',label ='Cluster3') # for third cluster
plt.scatter(x[y_predict==3,0],x[y_predict==3,1], s=100,c='cyan',label ='Cluster4') # for fourth cluster

plt.scatter(b.cluster_centers_[:,0],b.cluster_centers_[:,1],s = 300,c='yellow',label='centroid')
plt.title('Cluster of customers')
plt.xlabel("Annual Income(k$) ")
plt.ylabel("Spending Score(1-100) ")
plt.legend()
plt.show()
```

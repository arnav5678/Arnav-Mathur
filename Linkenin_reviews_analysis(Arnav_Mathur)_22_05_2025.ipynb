# **Adding Sentiments Labels in the Data**
We will use Textblob library. Textblob provides a polarity scores raining from -1(very negative) to 1(very positive) for a given text.We can use this score to classify each review's sentiment as positive,negative or neutral.

```python
from textblob import TextBlob
```

```python
def textblob_sentiment_analysis(review):
  sentiment = TextBlob(review).sentiment
  if sentiment.polarity > 0.1:
    return "Postitive"
  elif sentiment.polarity < -0.1:
    return 'Negative'
  else:
    return  "Neutral"
```

```python
df['Sentiment']= df['Review'].apply(textblob_sentiment_analysis)
```
